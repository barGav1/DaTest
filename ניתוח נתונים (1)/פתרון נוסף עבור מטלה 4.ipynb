```python
url = 'https://raw.githubusercontent.com/ShaiYona/Data-Science2021B/main/Assignments/4/nba.csv'
data= pd.read_csv(url)
```

# Q1 

```python
data_new = data.dropna().reset_index()
data_new = data[['Age','Height','Weight','Salary']]
data_new.head(3)
```

# Q2

```python
data_new['Height'] = data_new['Height'].astype('category').cat.codes 
data_new[['Age','Height','Weight','Salary']].head(3)
```

# Q3

```python
fig, axes = plt.subplots(figsize=(20, 5), ncols=2)
data['Position'].value_counts().plot.pie(ax=axes[0] ,autopct="%1.1f%%")
sns.countplot(ax=axes[1], x="Position", data=data)
plt.show()
```

# Q4

```python
features = ['Weight','Height','Age']
X = data_new[features]
y = data_new['Salary']
X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size=0.3, random_state=1)
X_train.head()
```
