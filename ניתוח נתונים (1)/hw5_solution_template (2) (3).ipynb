```python
data_rel = data[data['region'].str.contains('Orlando')]    
data_rel = data_rel[['Unnamed: 0','Date','AveragePrice','total_sold','small_sold','large_sold','sma','Total Bags','Small Bags', 'Large Bags', 'XLarge Bags','type','year','region']]
data_rel = data_rel[(data_rel.year==2016)]
data_rel.shape[0]
```

# Q4

In order to check the total avocado sold we used the "Datetime" function since they asked us for month 9 only, we did ".month" and then we were compared to month 9 only. We then did a groupby and put in the Total Bags and the month and then agreed with the ".sum" function
